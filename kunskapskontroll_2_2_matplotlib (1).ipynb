### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

En figur är behållaren på högsta nivån för allt i en tomt. Den rymmer en eller flera axlar (subplotter). Figuren är hela fönstret eller arbetsytan där alla plotelement (data, etiketter, förklaringar, etc.) återges.

Exempel: fig = plt.figure() skapar en figur.
 

En axel hänvisar till en enskild x-axel eller y-axel som hanterar bockmärken, etiketter och gränser för motsvarande dimension. Varje axelobjekt innehåller två axelobjekt (ett för x-axeln och ett för y-axeln).

Exempel: ax.set_xlim(0, 10) ställer in x-axelns gränser.
 

Axlar representerar området där data faktiskt plottas. En figur kan ha en eller flera axlar (delplots), och varje axel har sin egen x-axel och y-axel. Axlarna är ansvariga för att skala data och återge plotten.

Exempel: fig, ax = plt.subplots() skapar en figur med en enda axel.
 

artist är vilket visuellt element som helst som återges i handlingen. Detta inkluderar linjer, text (titlar, etiketter), markörer, bockar och till och med själva axlarna. I huvudsak är allt som ritas på figuren en konstnär.

Exempel: ax.plot(x, y) skapar en linjekonstnär.
 

### When plotting in Matplotlib, what is the expected input data type?

NumPy arrays or numpy.ma.masked_array
 

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Create an array of x values from -4 to 4
x = np.linspace(-4, 4, 100)
y = x ** 2
 
# Create a figure and axes
fig, ax = plt.subplots()
 
# Plot the function y = x^2
ax.plot(x, y)
 
# Set title and labels
ax.set_title('Plot of y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
 
# Show the plot
plt.show()
 
 
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
 
# Data for quadratic plot
x = np.linspace(-2, 2, 100)
y = x**2
 
# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
 
# Create a figure and two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns
 
# First subplot: Scatter plot
ax1.scatter(random_data_x, random_data_y, color='blue', alpha=0.5)
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
 
# Second subplot: Bar plot
ax2.bar(names, values, color='green')
ax2.set_title('Fruit Count')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Count')
 
# Adjust layout to prevent overlap
plt.tight_layout()
 
# Show the plot
plt.show()
 
```
